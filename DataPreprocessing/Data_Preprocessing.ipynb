```python
# Overview the movie meta dataset

import pandas as pd
import numpy as np
```

# Data Loading

```python
df = pd.read_csv("../data/movies_metadata.csv")
df.head()
```

# Data Viewing

```python
df.info()
```

```python
# drop non important features
# df.drop(["belongs_to_collection","homepage","tagline","overview","original_language"], axis=1,inplace=True)
```

```python
# Find the missing values for each column
df.isnull().sum()
```

```python
# Find the movies with missing title
print(df[df['title'].isnull()])
```

```python
# convert budget to float
df['budget'] = df['budget'].apply(pd.to_numeric, errors='coerce')
# drop movies with 0 budget
# df = df[df['budget'] != 0]
```

```python
# drop movies with missing title
df = df[df['title'].notna()]
# drop movies with missing revenue
# df = df[df['revenue'].notna()]
df.isnull().sum()
```

```python
#find movie with missing release data
# print(df[df['release_date'].isnull()])

# fill the release_data and runtime
# df["release_date"].fillna("2010-01-01",inplace=True)
# df["runtime"].fillna(df["runtime"].mean(),inplace=True)
# df.isnull().sum()
```

```python
# refill missing release date
df["release_date"].fillna("2010-01-01",inplace=True)
# fill runtime

df["runtime"].fillna(df["runtime"].mean(),inplace=True)
df.isnull().sum()
```

```python
# split release date to year and year
df["release_year"] = df["release_date"].apply(lambda x : int(x[:4]))
df["release_month"] = df["release_date"].apply(lambda x : x[5:7]).astype(int)
df = df[df["release_month"]!= ""]
df.head(5)
```

```python
df.info()
```

```python
# rename vote_average and vote_count to ratings and rating_count
df["rating"] = df["vote_average"]
df["rating_count"] = df["vote_count"]

df = df.drop("vote_average", 1)
df = df.drop("vote_count", 1)
df.head(5)
```

```python
# removing features which are irrelevant for our project
df.drop(["imdb_id", "poster_path","original_title"], axis=1,inplace=True)
```

# Export cleaned dataset

```python
df.to_csv("clean_movies_metadata.csv", index=False)
```
